```python
df=pd.DataFrame(iris.data,columns=iris.feature_names)
```

```python
df['result']=iris.target
```

```python
iris.target_names
```

```python
type=[]
for i,x in df.iterrows():
    if(x.result==0):
       type.append('setosa')
    elif(x.result==1):
       type.append('versicolor')
    else:
       type.append('verginica')
```

```python
# also by
df.result.apply(lambda x:iris.target_names[x])
```

```python
df['type']=type
```

```python
# TRAINING DATA MAKING
```

```python
from sklearn.model_selection import train_test_split
```

```python
X=df.iloc[:,0:4]
y=df.result
```

```python
x_train,x_test,y_train,y_test=train_test_split(X,y,test_size=0.2)
```

```python
# import SVC
```

```python
from sklearn.svm import SVC
```

```python
model=SVC()
```

```python
# TRAIN BY SVC
```

```python
model.fit(x_train,y_train)
```

```python
model.score(x_test,y_test)
```

```python
# TEST DATA
```

```python
predicted_y_test=model.predict(x_test)
```
